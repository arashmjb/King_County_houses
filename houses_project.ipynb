```python
# houses with price between 0.4 to 0.5 milion dollar that has at least 4 bedrooms and 2000 sqft_above and yr_built or renovated was after 2013
houses[(houses["price"].between(400000,500000)) & (houses["bedrooms"] > 3) & (houses["sqft_above"] >= 2000) & ((houses["yr_built"] > 2014) | (houses["yr_renovated"] >= 2014))]
```

```python
houses["price_per_sqft"] = houses["price"] / houses["sqft_lot"]
houses["price_per_sqft_living"] = houses["price"] / houses["sqft_living"]
```

```python
# top ten the most expensive houses based on price per sqft
houses.sort_values("price_per_sqft" , ascending=False).head(10)
```

```python
# top ten the most expensive houses based on price per sqft living
houses.sort_values("price_per_sqft_living" , ascending=False).head(10)
```

```python
# which zipcode of df has the top 50 houses based on price per sqft living 
houses.sort_values("price_per_sqft_living" , ascending=False).head(50)["zipcode"].value_counts().plot(kind="bar")
```
